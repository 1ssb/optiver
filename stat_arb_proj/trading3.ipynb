```python
import pandas as pd
import numpy as np

df = pd.read_csv('rdata.csv')
df.head()
df.columns

df = df[[
    'TIMESTAMP',
    'FEEDCODE_HSI',
    'FEEDCODE_HHI',
    'LAST_BID_PRICE_HSI',
    'LAST_ASK_PRICE_HSI',
    'LAST_BID_VOLUME_HSI',
    'LAST_ASK_VOLUME_HSI',
    'LAST_BID_PRICE_HHI',
    'LAST_ASK_PRICE_HHI',
    'LAST_BID_VOLUME_HHI',
    'LAST_ASK_VOLUME_HHI'
]]

df
```

```python
# Calculate midpoint prices
df['MIDPOINT_HSI'] = (df['LAST_BID_PRICE_HSI'] + df['LAST_ASK_PRICE_HSI']) / 2
df['MIDPOINT_HHI'] = (df['LAST_BID_PRICE_HHI'] + df['LAST_ASK_PRICE_HHI']) / 2

# Calculate log returns
df['LOG_RETURN_HSI'] = np.log(df['MIDPOINT_HSI'] / df['MIDPOINT_HSI'].shift(1))
df['LOG_RETURN_HHI'] = np.log(df['MIDPOINT_HHI'] / df['MIDPOINT_HHI'].shift(1))

# Drop NaN values caused by the shift
df = df.dropna()
```

```python
# Apply EWMA smoothing with an expanded window (e.g., 30 minutes) for both HSI and HHI
df['LOG_RETURN_HSI_SMOOTHED_EWMA'] = df['LOG_RETURN_HSI'].ewm(span=30, adjust=False).mean()
df['LOG_RETURN_HHI_SMOOTHED_EWMA'] = df['LOG_RETURN_HHI'].ewm(span=30, adjust=False).mean()

# Drop NaN values caused by the window
df = df.dropna()
```

```python
# Calculate rolling statistics
window = 60  # Set rolling window size (e.g., 60 minutes)
df['ROLLING_STD_HSI'] = df['LOG_RETURN_HSI'].rolling(window=window).std()
df['ROLLING_STD_HHI'] = df['LOG_RETURN_HHI'].rolling(window=window).std()
df['ROLLING_CORR'] = df['LOG_RETURN_HSI'].rolling(window=window).corr(df['LOG_RETURN_HHI'])

# Compute rolling hedge ratio
df['HEDGE_RATIO'] = (df['ROLLING_CORR'] * df['ROLLING_STD_HHI']) / df['ROLLING_STD_HSI']

# Generate hedged returns
df['HEDGED_RETURN'] = df['LOG_RETURN_HSI'] - df['HEDGE_RATIO'] * df['LOG_RETURN_HHI']

# Smooth the hedged returns
df['HEDGED_RETURN_SMOOTHED'] = df['HEDGED_RETURN'].ewm(span=window).mean()

# Drop NaN values caused by rolling calculations
df = df.dropna()
```

```python
# Define rolling window size for calculating Z-scores
z_window = 60  # e.g., 60 minutes

# Calculate rolling mean and standard deviation for hedged returns
df['HEDGED_RETURN_MEAN'] = df['HEDGED_RETURN_SMOOTHED'].rolling(window=z_window).mean()
df['HEDGED_RETURN_STD'] = df['HEDGED_RETURN_SMOOTHED'].rolling(window=z_window).std()

# Calculate Z-score
df['Z_SCORE'] = (df['HEDGED_RETURN_SMOOTHED'] - df['HEDGED_RETURN_MEAN']) / df['HEDGED_RETURN_STD']

# Drop NaN values caused by rolling calculations
df = df.dropna()
```

```python
# def generate_signals(df, entry_threshold=2, exit_threshold=0, credit_limit=10, delta_risk=1):
#     """
#     Generate trading signals with a credit-limited position formulation.
    
#     Parameters:
#         df (DataFrame): Input DataFrame with Z-scores and other metrics.
#         entry_threshold (float): Z-score threshold for entering trades.
#         exit_threshold (float): Z-score threshold for exiting trades.
#         credit_limit (float): Maximum allowed cumulative position size (in units).
#         delta_risk (float): Growth parameter for incremental position risk.
#     """
#     # Reset index to ensure proper alignment
#     df = df.reset_index(drop=True)

#     # Initialize variables
#     signals = []
#     prev_hsi_signal = 0
#     prev_hhi_signal = 0
#     current_credit = 0  # Tracks the current credit usage

#     # Iterate over rows using iterrows
#     for index, row in df.iterrows():
#         z_score = row['Z_SCORE']  # Access Z_SCORE directly from row
#         hsi_signal = 0
#         hhi_signal = 0

#         # Entry signals with credit limit check
#         if z_score > entry_threshold and current_credit + delta_risk <= credit_limit:
#             hsi_signal = -1  # Short HSI
#             hhi_signal = 1   # Long HHI
#             current_credit += delta_risk
#         elif z_score < -entry_threshold and current_credit + delta_risk <= credit_limit:
#             hsi_signal = 1   # Long HSI
#             hhi_signal = -1  # Short HHI
#             current_credit += delta_risk

#         # Exit signals
#         elif abs(z_score) < exit_threshold:
#             hsi_signal = 0
#             hhi_signal = 0
#             current_credit = max(0, current_credit - delta_risk)  # Reduce credit

#         # Carry over previous signal if no new signal
#         if hsi_signal == 0:
#             hsi_signal = prev_hsi_signal
#         if hhi_signal == 0:
#             hhi_signal = prev_hhi_signal

#         # Append the signals and update previous signals
#         signals.append((hsi_signal, hhi_signal))
#         prev_hsi_signal = hsi_signal
#         prev_hhi_signal = hhi_signal

#     # Add signals to DataFrame
#     df['HSI_SIGNAL'], df['HHI_SIGNAL'] = zip(*signals)

#     return df


def generate_signals(
    df, entry_threshold=2, exit_threshold=0, 
    position_limit=500
):
    """
    Generate trading signals with strict cumulative volume caps.
    
    Parameters:
        df (DataFrame): Input DataFrame with Z-scores and other metrics.
        entry_threshold (float): Z-score threshold for entering trades.
        exit_threshold (float): Z-score threshold for exiting trades.
        volume_limit (int): Maximum allowable cumulative signed volume for either instrument.
    """
    """
    Generate trading signals with cumulative position limits.
    
    Parameters:
        df (DataFrame): Input DataFrame with Z-scores and other metrics.
        entry_threshold (float): Z-score threshold for entering trades.
        exit_threshold (float): Z-score threshold for exiting trades.
        position_limit (int): Maximum allowable cumulative net position for either instrument.
    """
    # Reset index to ensure proper alignment
    df = df.reset_index(drop=True)

    # Initialize variables
    signals = []
    cumulative_hsi_position = 0  # Tracks net cumulative HSI position
    cumulative_hhi_position = 0  # Tracks net cumulative HHI position

    # Iterate over rows using iterrows
    for index, row in df.iterrows():
        z_score = row['Z_SCORE']  # Access Z_SCORE directly from row
        hsi_signal = 0
        hhi_signal = 0

        # Entry signals with cumulative position limits
        if z_score > entry_threshold:
            # Short HSI and Long HHI only if within cumulative position limits
            if cumulative_hsi_position - 1 >= -position_limit and cumulative_hhi_position + 1 <= position_limit:
                hsi_signal = -1  # Short HSI
                hhi_signal = 1   # Long HHI
        elif z_score < -entry_threshold:
            # Long HSI and Short HHI only if within cumulative position limits
            if cumulative_hsi_position + 1 <= position_limit and cumulative_hhi_position - 1 >= -position_limit:
                hsi_signal = 1   # Long HSI
                hhi_signal = -1  # Short HHI

        # Exit signals
        elif abs(z_score) < exit_threshold:
            # Reset positions to exit trades
            hsi_signal = 0
            hhi_signal = 0

        # Update cumulative positions
        cumulative_hsi_position += hsi_signal
        cumulative_hhi_position += hhi_signal

        # Enforce cumulative position limits (revert signals if exceeded)
        if abs(cumulative_hsi_position) > position_limit:
            cumulative_hsi_position -= hsi_signal
            hsi_signal = 0
        if abs(cumulative_hhi_position) > position_limit:
            cumulative_hhi_position -= hhi_signal
            hhi_signal = 0

        # Append the signals
        signals.append((hsi_signal, hhi_signal))

    # Add signals to DataFrame
    df['HSI_SIGNAL'], df['HHI_SIGNAL'] = zip(*signals)

    return df

# Apply the signal generation function
df = generate_signals(df)
```

```python
# Generate trades for HSI
hsi_trades = df[df['HSI_SIGNAL'] != 0].copy()
hsi_trades['FEEDCODE'] = hsi_trades['FEEDCODE_HSI']
hsi_trades['BUYSELL'] = np.where(hsi_trades['HSI_SIGNAL'] == 1, 'B', 'S')
hsi_trades['PRICE'] = np.where(
    hsi_trades['HSI_SIGNAL'] == 1, hsi_trades['LAST_ASK_PRICE_HSI'], hsi_trades['LAST_BID_PRICE_HSI']
)
hsi_trades['VOLUME'] = np.where(
    hsi_trades['HSI_SIGNAL'] == 1, hsi_trades['LAST_ASK_VOLUME_HSI'], hsi_trades['LAST_BID_VOLUME_HSI']
)
hsi_trades = hsi_trades[['TIMESTAMP', 'FEEDCODE', 'BUYSELL', 'PRICE', 'VOLUME']]

# Generate trades for HHI
hhi_trades = df[df['HHI_SIGNAL'] != 0].copy()
hhi_trades['FEEDCODE'] = hhi_trades['FEEDCODE_HHI']
hhi_trades['BUYSELL'] = np.where(hhi_trades['HHI_SIGNAL'] == 1, 'B', 'S')
hhi_trades['PRICE'] = np.where(
    hhi_trades['HHI_SIGNAL'] == 1, hhi_trades['LAST_ASK_PRICE_HHI'], hhi_trades['LAST_BID_PRICE_HHI']
)
hhi_trades['VOLUME'] = np.where(
    hhi_trades['HHI_SIGNAL'] == 1, hhi_trades['LAST_ASK_VOLUME_HHI'], hhi_trades['LAST_BID_VOLUME_HHI']
)
hhi_trades = hhi_trades[['TIMESTAMP', 'FEEDCODE', 'BUYSELL', 'PRICE', 'VOLUME']]

# Combine HSI and HHI trades
trade_table = pd.concat([hsi_trades, hhi_trades], ignore_index=True)

# Handle aggregation for same TIMESTAMP and FEEDCODE
trade_table['PRICE_VOLUME'] = trade_table['PRICE'] * trade_table['VOLUME']
aggregated_table = (
    trade_table.groupby(['TIMESTAMP', 'FEEDCODE', 'BUYSELL'], as_index=False)
    .agg(
        VOLUME=('VOLUME', 'sum'),
        PRICE_VOLUME=('PRICE_VOLUME', 'sum')
    )
)
aggregated_table['PRICE'] = aggregated_table['PRICE_VOLUME'] / aggregated_table['VOLUME']

# Drop intermediate calculation column
trades = aggregated_table[['TIMESTAMP', 'FEEDCODE', 'BUYSELL', 'PRICE', 'VOLUME']]
```
